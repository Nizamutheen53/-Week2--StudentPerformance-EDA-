```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# For better visuals
sns.set(style="whitegrid")
plt.rcParams['figure.figsize'] = (12, 6)

# Load the dataset
df = pd.read_csv('country_wise_latest.csv')

# Display top 5 rows
print("Top 5 rows of the dataset:")
display(df.head())
```

```python
# Check dataset information
print("\nDataset Information:")
df.info()

# Summary statistics
print("\nSummary Statistics:")
display(df.describe())

# Check for null values
print("\nNull values in the dataset:")
print(df.isnull().sum())
```

```python
# Check for duplicates
print("\nNumber of duplicate rows:", df.duplicated().sum())

# Number of unique countries
print("\nTotal number of countries in the dataset:", df['Country/Region'].nunique())

# Most affected countries (by confirmed cases)
print("\nTop 10 most affected countries by confirmed cases:")
top_affected = df.sort_values('Confirmed', ascending=False)[['Country/Region', 'Confirmed', 'Deaths', 'Recovered']].head(10)
display(top_affected)
```

```python
# Line chart for confirmed cases over countries (since this dataset is latest, not over time)
# Simulating a line chart with sorted countries
top_confirmed = df.sort_values('Confirmed', ascending=False).head(15)
plt.figure(figsize=(14, 6))
sns.lineplot(data=top_confirmed, x='Country/Region', y='Confirmed', marker='o')
plt.title('Confirmed COVID-19 Cases - Top 15 Countries')
plt.xticks(rotation=45)
plt.show()
```

```python
# Bar chart showing top 10 countries with highest deaths
top_deaths = df.sort_values('Deaths', ascending=False).head(10)
plt.figure(figsize=(14, 6))
sns.barplot(data=top_deaths, x='Country/Region', y='Deaths', palette='Reds_r')
plt.title('Top 10 Countries with Highest COVID-19 Deaths')
plt.xticks(rotation=45)
plt.show()
```

```python
# Pie chart of global recovered vs deaths
total_recovered = df['Recovered'].sum()
total_deaths = df['Deaths'].sum()
plt.figure(figsize=(7, 7))
plt.pie([total_recovered, total_deaths], labels=['Recovered', 'Deaths'],
        colors=['#66b3ff', '#ff6666'], autopct='%1.1f%%', startangle=140)
plt.title('Global Recovered vs Deaths')
plt.show()
```

```python
# Short Observations
print("\n--- Observations ---")
print(f"Total number of countries: {df['Country/Region'].nunique()}")
print(f"Total confirmed cases globally: {df['Confirmed'].sum():,}")
print(f"Total recovered globally: {total_recovered:,}")
print(f"Total deaths globally: {total_deaths:,}")
print(f"Most affected country: {df.sort_values('Confirmed', ascending=False).iloc[0]['Country/Region']}")
```
